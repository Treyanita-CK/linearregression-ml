```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

data = 'C:\\Users\\Treyanita Citra K\\Python\\Latihan Machine Learning\\File\\data_car_price.csv'
df = pd.read_csv(data)
df.head()
```

```python
import matplotlib.pyplot as plt

# melihat perbesaran
f = plt.figure(figsize=(12, 4))

f.add_subplot(1, 2, 1)
df['horsepower'].plot(kind='kde')

f.add_subplot(1, 2, 2)
plt.boxplot(df['horsepower'])

plt.show()
```

```python
plt.scatter(df['horsepower'], df['price'])
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.title('Scatter Plot Horsepower vs Price')
plt.show()
```

```python
df.corr()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# identifikasi fitur dan label

x = df['horsepower'].values.reshape(-1, 1)
y = df['price'].values.reshape(-1, 1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)

# buat objek linear regression
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)

# print lin_reg coef (m) atau gradien
# print lin_reg intercept (b) atau bias

print("Hasil dari koefesion linear regression m: ", lin_reg.coef_)
print("Hasil dari bias linear regression b : ", lin_reg.intercept_)
```

```python
lin_reg.score(x_test, y_test)
```

```python
#Ketujuh, visualisasi aktual vs prediksi menggunakan data testing
#Garis merah merupakan garis regresi dari persamaan yang kita dapat tadi
y_prediksi = lin_reg.predict(x_test)
plt.scatter(x_test, y_test)
plt.plot(x_test, y_prediksi, c='r')
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.title('Plot Horsepower vs Price')
```

```python
#Prediksi harga mobil dengan horsepower 100
lin_reg.predict([[100]])
```
